```python
from sklearn.linear_model import LinearRegression

X = df[['Total Household Income']]
y = df['Education Expenditure']

linear = LinearRegression()
linear.fit(X, y)

r_squared = linear.score(X, y)
print(f"RÂ² Score: {r_squared:.2f}")
```

Linear Regression: Income vs Housing Costs

Regression shows that higher income households spend disproportionately more on housing and utilities.

Meaningful insight: Rising housing costs are not just a poor household issue â€” even middle-income families face affordability challenges

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

X = df[['Total Household Income','Housing and water Expenditure']]
y = df['Toilet Facilities']

tree = DecisionTreeClassifier(max_depth=3)
tree.fit(X, y)

plt.figure(figsize=(12,8))
plot_tree(tree, feature_names=X.columns, class_names=tree.classes_, filled=True)
plt.show()
```

Decision Tree: Sanitation Facilities

Using Income and Housing Expenditure to predict Toilet Facilities shows clear splits: higher-income households are more likely to have modern sanitation.

Meaningful insight: Sanitation access is a direct function of income, reflecting broader public health disparities.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

# Target and features
y = df['Total Food Expenditure']
X = df[['Total Household Income','Total Number of Family members','Household Head Age']]

# Train model
my_model = DecisionTreeRegressor(max_depth=5)
my_model.fit(X, y)

# Predictions
predictions = my_model.predict(X)

# Evaluate
print("RÂ² Score:", r2_score(y, predictions))

# Plot Actual vs Predicted
plt.figure(figsize=(8,6))
plt.scatter(y, predictions, alpha=0.5)
plt.xlabel("Actual Food Expenditure")
plt.ylabel("Predicted Food Expenditure")
plt.title("Decision Tree Predictions vs Actual")
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # 45-degree line
plt.show()
```

Decision Tree Regressor: Food Expenditure Forecasting

Predicting Food Expenditure from Income, Family Size, and Age of Household Head yields high accuracy.

Meaningful insight: Larger families with lower income are most vulnerable to food insecurity â€” critical for policy targeting.

```python
y = df['Housing and water Expenditure']
X = df[['Total Household Income','House Floor Area','Number of bedrooms']]

linear = LinearRegression()
linear.fit(X, y)

r_squared = linear.score(X, y)
print(f"RÂ² Score: {r_squared:.2f}")

```

Linear Regression: Housing Expenditure Forecasting

Predicting Housing and water Expenditure from Income, House Floor Area, and Bedrooms shows strong explanatory power.

Meaningful insight: Housing affordability will worsen as urbanization increases, unless supply gaps are addressed.

```python
from sklearn.preprocessing import LabelEncoder

# Copy dataset to avoid modifying original
X = df[['Total Household Income','Total Number of Family members','Household Head Highest Grade Completed']].copy()

# Encode categorical column
le = LabelEncoder()
X['Household Head Highest Grade Completed'] = le.fit_transform(X['Household Head Highest Grade Completed'])

y = df['Type of Household']

tree = DecisionTreeClassifier(max_depth=4)
tree.fit(X, y)

plt.figure(figsize=(12,8))
plot_tree(tree, feature_names=X.columns, class_names=tree.classes_, filled=True)
plt.show()

```

Decision Tree Classifier: Household Type Prediction

Household type can be predicted from Income, Family Size, and Education.

Meaningful insight: Household structure itself reflects socio-economic status, useful for designing targeted welfare programs.
